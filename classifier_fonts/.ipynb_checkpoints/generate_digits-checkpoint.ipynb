# Generate digit images 

The final goal of this notebook is to create a folder that will contain digit images.

First we will explore how to write a digit, and finally we will generate images of digits.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Write a digit on an image

To do so we use the `putText()` function from `cv2` : see https://www.geeksforgeeks.org/python-opencv-cv2-puttext-method/

```python
#create an empty image (black image)
img = np.zeros((480,640))

#use cv2 to write a digit
cv2.putText(img, str(3) ,(230,320),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX , 10, (255,255,255),15,cv2.LINE_AA)

#show the image (we use cmap=gray because the image is in greyscale (not RGB))
plt.imshow(img, cmap="gray", vmin=0, vmax=255)
cv2.imwrite('test1.jpg', img)
```

There is different fonts that can be used to write a digit. Therefore, we can use the different fonts as a way to make the classification model more robust to different ways of drawing a digit

# Rotate and translate the digit

In next sections, we will use the rotation and translation operations on images. Here is an example.

```python
#create an empty image (black image)
img = np.zeros((480,640))

#use cv2 to write a digit
cv2.putText(img, str(4) ,(230,320),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX , 10, (255,255,255),15,cv2.LINE_AA)

#rotate the image
rows,cols = img.shape
M = cv2.getRotationMatrix2D((cols/2,rows/2),10,1) #rotation of 10 degress
rot = cv2.warpAffine(img,M,(cols,rows))

#translate the image
M = np.float32([[1,0,100],[0,1,50]]) #for a (100,50) translation
trans = cv2.warpAffine(img,M,(cols,rows))

#show the image (we use cmap=gray because the image is in greyscale (not RGB))
fig, axs = plt.subplots(1,3, figsize=(13,5))
axs[0].imshow(img, cmap="gray", vmin=0, vmax=255)
axs[0].set_title('Original image')

axs[1].imshow(rot, cmap="gray", vmin=0, vmax=255)
axs[1].set_title('Rotation')

axs[2].imshow(trans, cmap="gray", vmin=0, vmax=255)
axs[2].set_title('Translation');

```

A challenging problem is the location of the digit : imagine all the ones are located at the top left of the drawing window, and other digits are located somewhere else. Thus the classifier may learn that every symbol at the top left of the image are ones. Therefore we want to be insensitive to the spatial position of the image.

There is the same phenomenon with the way the digit is drawn. If every ones used to train the classifier are perfectly vertical, a person's drawn one that is not purely vertical (which is expected) will maybe not be recognized as a one. Thus we want to be insensitive to the rotation of the image.

## Zoom on the digit

To be insensitive to the location of the digit, we can center it : we remove all the black pixels around the digit

```python
#exemple of centering a digit
img = np.zeros((480,640))
#use cv2 to write a digit
cv2.putText(img, str(0) ,(230,320),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX , 10, (255,255,255),15,cv2.LINE_AA)

rows, cols = np.where(img==255) #we get rows and cols 
#now we find the corners by finding the max and min of the previous arrays 'rows' and 'cols'
img_centered = img[min(rows):max(rows), min(cols):max(cols)]

fig, axs = plt.subplots(1,2, figsize=(10,5))
axs[0].imshow(img, cmap="gray", vmin=0, vmax=255)
axs[0].set_title('Original image')

axs[1].imshow(img_centered, cmap="gray", vmin=0, vmax=255)
axs[1].set_title('Centered image')
```

Center the image change the shape of the image. This shape vary with the concerned digit :

```python
for i in range(10):
    img = np.zeros((480,640))

    #use cv2 to write a digit
    cv2.putText(img, str(i) ,(230,320),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX , 10, (255,255,255),15,cv2.LINE_AA)
    
    rows, cols = np.where(img==255)
    img_centered = img[min(rows):max(rows), min(cols):max(cols)]
   
    print('digit : {} ----- Shape : {}'.format(i, img_centered.shape))
```

The digit 1 has a centered shape much different from other. All the images must have the same shape, so we need to reshape the images. Since almost all  shapes are around (226, 156), we try to respect this ratio by using a shape of (44, 28)

```python
#exemple of reshaping
L=[]
for i in range(10):
    img = np.zeros((480,640))

    #use cv2 to write a digit
    cv2.putText(img, str(i) ,(230,320),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX , 10, (255,255,255),15,cv2.LINE_AA)
    
    rows, cols = np.where(img==255)
    img_centered = img[min(rows):max(rows), min(cols):max(cols)]
    img = cv2.resize(img_centered, (28, 44))
    
    L.append(img)
    
    
fig, axs = plt.subplots(1,3, figsize=(15,5))
axs[0].imshow(L[0], cmap="gray", vmin=0, vmax=255)
axs[1].imshow(L[1], cmap="gray", vmin=0, vmax=255)
axs[2].imshow(L[2], cmap="gray", vmin=0, vmax=255)
```

## Explore the fonts

There is 8 fonts in total, we show them below. The idea is to use different fonts to train the classifier, therefore making the classifier more robust to different writings.

```python
#test : show all fonts on a single image
img = np.zeros((480,640))
w, h = img.shape
fonts = [cv2.FONT_HERSHEY_SIMPLEX, cv2.FONT_HERSHEY_PLAIN, cv2.FONT_HERSHEY_DUPLEX, cv2.FONT_HERSHEY_COMPLEX,
         cv2.FONT_HERSHEY_TRIPLEX, cv2.FONT_HERSHEY_COMPLEX_SMALL, cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,
        cv2.FONT_HERSHEY_SCRIPT_COMPLEX]
for i in range(10):
    h=30
    w = 10+ 50*i
    for font in fonts:
        cv2.putText(img, str(i) ,(w,h),font , 1, (255,255,255),2,cv2.LINE_AA)
        h+= 60
        
plt.figure(figsize=(10,7))
plt.imshow(img, cmap="gray", vmin=0, vmax=255)
```

## Create the training digit images

In this part, we will create the folder containing the data. The data will be images of digits from 0 to 9. 

We will create 77 images for each digit. Thus we will have 770 images in total. To increase the robustness, we will use 7 different fonts (we keep one font for the testing data). We will also rotate the digits.

Therefore for a given digit, and a given font, we will create one image for the cases where the digit is rotated by a value of **0°, 2°, 4°, 6°, 8°, 10°, -2°, ..., -10°**.


Therefore we will have 11 images for a given digit and a given font.

Depending on the results of the machine learning model, we can increase the number of images.

```python
#create the training images
import os
if not os.path.exists('data'):
    os.makedirs('data')
    
if not os.path.exists('data/train'):
    os.makedirs('data/train')
    
    
fonts = [cv2.FONT_HERSHEY_SIMPLEX, cv2.FONT_HERSHEY_PLAIN, cv2.FONT_HERSHEY_COMPLEX, cv2.FONT_HERSHEY_TRIPLEX,
         cv2.FONT_HERSHEY_COMPLEX_SMALL, cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, cv2.FONT_HERSHEY_SCRIPT_COMPLEX]
# we keep 'cv2.FONT_HERSHEY_DUPLEX' for the testing data

rotations = [0, 2,4,6,8,10,-2,-4,-6,-8,-10]


for i in range(10):
    count=0
    for font in fonts:
        for rotation in rotations:
            #create an empty image (black image)
            img = np.zeros((480,640))

            #use cv2 to write a digit
            cv2.putText(img, str(i) ,(230,320),font , 10, (255,255,255),15,cv2.LINE_AA)
            
            #rotate the image
            rows,cols = img.shape
            M = cv2.getRotationMatrix2D((cols/2,rows/2),rotation,1)
            rot = cv2.warpAffine(img,M,(cols,rows))
            
            #center the digit and resize it
            rows, cols = np.where(rot==255)
            img_centered = rot[min(rows):max(rows), min(cols):max(cols)]
            img = cv2.resize(img_centered, (28, 44))
            
            #save th image in the folder (we save 2 times the same image) 
            #file path (we use the format '0_3' for the 3rd image of the digit 0)
            file_path = 'data/train/'+str(i)+'_'+str(count)+'.jpg'
            cv2.imwrite(file_path, img)
            count+=1
```

## Create the testing digit images

We also need to create the testing images. We will use 110 images (to have around 85/15 ratio between training/testing data), 11 per digit.

We will use the same rotations

We will use the last font : `cv2.FONT_HERSHEY_DUPLEX`.

```python
if not os.path.exists('data'):
    os.makedirs('data')
    
if not os.path.exists('data/test'):
    os.makedirs('data/test')
    
font = cv2.FONT_HERSHEY_DUPLEX
rotations = [0, 2,4,6,8,10,-2,-4,-6,-8,-10]


for i in range(10):
    count=0
    for rotation in rotations:
        #create an empty image (black image)
        img = np.zeros((480,640))

        #use cv2 to write a digit
        cv2.putText(img, str(i) ,(230,320),font , 10, (255,255,255),15,cv2.LINE_AA)
        
        #rotate the image
        rows,cols = img.shape
        M = cv2.getRotationMatrix2D((cols/2,rows/2),rotation,1)
        rot = cv2.warpAffine(img,M,(cols,rows))
        
        #center the digit and resize it
        rows, cols = np.where(rot==255)
        img_centered = rot[min(rows):max(rows), min(cols):max(cols)]
        img = cv2.resize(img_centered, (28, 44))
            
        #save the image in the folder
        #file path (we use the format '0_3' for the 3rd image of the digit 0)
        file_path = 'data/test/'+str(i)+'_'+str(count)+'.jpg'
        cv2.imwrite(file_path, img)
        count+=1
```

## Create the train/test dataframes

To train the classifier, we need to create a dataframe that contains the name of the images as well as the label.

```python
train = pd.DataFrame(columns =['name', 'label'])
for i in range(10): #for each digit
    for j in range(77): #there is 77 images per digit
        file_name = str(i) + '_' + str(j) + '.jpg'
        train = train.append({'name': file_name, 'label' : i}, ignore_index=True)
        
print('Shape of dataframe : {}, unique labels : {}'.format(train.shape, train['label'].unique()))
train.head()
train.to_csv('data/train.csv', index=False)
```

We do the same for the testing data

```python
test = pd.DataFrame(columns =['name', 'label'])
for i in range(10): #for each digit
    for j in range(11): #there is 11 images per digit
        file_name = str(i) + '_' + str(j) + '.jpg'
        test = test.append({'name': file_name, 'label' : i}, ignore_index=True)
        
print('Shape of dataframe : {}, unique labels : {}'.format(test.shape, test['label'].unique()))
test.head()
test.to_csv('data/test.csv', index=False)
```
